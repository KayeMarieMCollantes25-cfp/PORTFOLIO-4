# Program: Student Budget Tracker
##### As a student, this program can help calculate exactly where allowance goes based on daily life. This can also help to continuously checks the remaining cash after every single entry. 

```python
##### Printing budget status
def show_status(money):
    print("Wallet Balance: PHP " + str(money))

print("---- Student Budget Tracker ----")

# Starting for variables
wallet = input("Enter the money: ")
wallet = float(wallet)
total_spent = 0.00
largest_expense = None

# Using while loop
keep_running = True
while keep_running: 
    show_status(wallet)
    print("")
    print("Options: 1 = Add Expense, 2 = View Summary, 3 = Exit")
    choice = input("Enter the option: ")

    if choice == "2":
        print("Checking your history...")
        continue 
    
    elif choice == "3": 
         print("Closing budget tracker.")
         break

    # Processing the expense
    expense_item = input("What did you buy?")
    cost = input("How much did it cost? PHP ")
    cost = float(cost)

    # Finding the largest expense
    if largest_expense is None or cost > largest_expense:
        largest_expense = cost

    wallet = wallet - cost
    total_spent  =  total_spent + cost
    print("Spent PHP " + str(cost) + " on " + str(expense_item))

    # Warning if the allowance is not enoug anymore
    if wallet < 50.0:
        print("WARNING!!! Your balance is below PHP 50.00. Stop spending!")


    print("-------------------------------------------------------------------")

# Using definite loop to get the result
print("")
print("---- Weekly Saving Tips ----")
tips = ["Bring a water tumbler", "Walk from School", "Cook meals at home"]

for item in tips: 
    print("-  " + str(item))

# Showing final result
print("")
print("---- Final Budget Summary ----")
print("Total Money Spent: PHP " + str(total_spent))
print("Your Largest Single Expense: PHP " + str(largest_expense))
```
